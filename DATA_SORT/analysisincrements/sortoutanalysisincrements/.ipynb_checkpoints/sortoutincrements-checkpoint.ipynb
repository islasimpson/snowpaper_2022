```python
import importlib
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import timedelta
import sys
```

```python
#ystart=1979
#yend=2014
ystart=2021
yend=2021
```

```python
analysis_18h_path = "/project/haggis/ERA5/forincrements/analysis/t2m/18h/"
analysis_06h_path = "/project/haggis/ERA5/forincrements/analysis/t2m/06h/"
forecast_06h_path = "/project/haggis/ERA5/forincrements/forecast/t2m/06h_step12/"
forecast_18h_path = "/project/haggis/ERA5/forincrements/forecast/t2m/18h_step12/"
outpath="/project/haggis/ERA5/forincrements/increments/t2m/"
```

```python
print(forecastday.time)
```

```python
for iyear in np.arange(ystart,yend+1,1):
    print(iyear)
    
    ### forecast 06h, step 12 and analysis at 18h of the same day
    analysis1 = xr.open_dataset(analysis_18h_path+'t2m_'+str(iyear)+'.nc')
    forecast1 = xr.open_dataset(forecast_06h_path+'tas_fc_06h_step12_'+str(iyear)+'.nc')
    forecast1['time'] = forecast1.time + np.timedelta64(12,'h')
    
    ### making sure the coordinates are the same
    forecast1['lon'] = analysis1.lon
    forecast1['lat'] = analysis1.lat
    
    increment1 = analysis1 - forecast1
    
    ### forecast 18h, step 12 and analysis at 06h of the following day
    analysis2 = xr.open_dataset(analysis_06h_path+'t2m_'+str(iyear)+'.nc')
    forecast2 = xr.open_dataset(forecast_18h_path+'tas_fc_18h_step12_'+str(iyear)+'.nc')
    forecast2['time'] = forecast2.time + np.timedelta64(12,'h')
    
    if (iyear == ystart):
        # drop January first of the analysis for 06h since we don't have an accompanying forecast
        analysis2 = analysis2.where(analysis2.time.dt.dayofyear != 1, drop=True)
        
    ### drop the last forecast and save it for later since it's part of the next year.
    if (iyear != ystart):
        forecast2 = xr.concat([forecastnextyear, forecast2], dim='time')
    
    forecastnextyear = forecast2.where(forecast2.time.dt.year == iyear+1, drop=True)
    forecast2 = forecast2.where(forecast2.time.dt.year != iyear+1, drop=True)
    
    forecast2['lon'] = analysis2.lon
    forecast2['lat'] = analysis2.lat
    
    increment2 = analysis2 - forecast2
    
    ## combine the two parts of the day
    increment = xr.concat([increment1, increment2], dim='time')
    analysis = xr.concat([analysis1, analysis2], dim='time')
    forecast = xr.concat([forecast1, forecast2], dim='time')
    
    ##sort based on time (not sure this is necessary)
    increment = increment.sortby('time', ascending=True)
    analysis = analysis.sortby('time', ascending=True)
    forecast = forecast.sortby('time', ascending=True)
    
    timeday = increment.time.groupby('time.dayofyear').mean('time')
    timeday = timeday.rename({'dayofyear':'time'})
    incrementday = increment.groupby('time.dayofyear').mean('time')
    incrementday = incrementday.rename({'dayofyear':'time'})
    incrementday['time'] = timeday
    incrementday = incrementday.rename({'t2m':'increment'})
    
    analysisday = analysis.groupby('time.dayofyear').mean('time')
    analysisday = analysisday.rename({'dayofyear':'time'})
    analysisday['time'] = timeday
    analysisday = analysisday.rename({'t2m':'analysis'})
    
    forecastday = forecast.groupby('time.dayofyear').mean('time')
    forecastday = forecastday.rename({'dayofyear':'time'})
    #forecastday = forecastday.isel(time=slice(0,forecastday.time.size-1))
    forecastday['time'] = timeday
    forecastday = forecastday.rename({'t2m':'forecast'})
    
    #sys.exit()
    
    incrementday.to_netcdf(outpath+'/increment_'+str(iyear)+'.nc')
    analysisday.to_netcdf(outpath+'/increment_'+str(iyear)+'.nc', mode='a')
    forecastday.to_netcdf(outpath+'/increment_'+str(iyear)+'.nc', mode='a')
```

```python
analysis1plot = analysis1.t2m.sel(lon=82.3, lat=66.6, method='nearest')
forecast1plot = forecast1.t2m.sel(lon=82.3, lat=66.6, method='nearest')
analysis2plot = analysis2.t2m.sel(lon=82.3, lat=66.6, method='nearest')
forecast2plot = forecast2.t2m.sel(lon=82.3, lat=66.6, method='nearest')
```

```python
fig = plt.figure(figsize=(16,16))

ax = fig.add_axes([0.1,0.8,0.4,0.18])
ax.set_xlim(0,90)

ax.plot(analysis1plot, label='analysis1')
ax.plot(forecast1plot, label='forecast1')
#ax.plot(analysis2plot, label='analysis2')
#ax.plot(forecast2plot, label='forecast2')
ax.legend()

ax= fig.add_axes([0.6,0.8,0.18,0.18])

ax.plot(analysis1plot.isel(time=slice(0,90)), forecast1plot.isel(time=slice(0,90)), "o")
ax.plot([220,270],[220,270])


ax = fig.add_axes([0.1,0.6,0.4,0.18])
ax.set_xlim(0,90)

ax.plot(analysis2plot, label='analysis2')
ax.plot(forecast2plot, label='forecast2')
ax.legend()


ax= fig.add_axes([0.6,0.6,0.18,0.18])
ax.plot(analysis2plot.isel(time=slice(300,364)), forecast2plot.isel(time=slice(300,364)), "o")
ax.plot([220,270],[220,270])
```
