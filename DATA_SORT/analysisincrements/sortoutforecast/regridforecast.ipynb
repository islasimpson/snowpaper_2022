```python
import importlib
import sys
import xarray as xr
import numpy as np
import xesmf as xe
```

```python
#ystart=1979
#yend=2014
ystart=2021
yend=2021
inpath="/project/haggis/ERA5/download/tas_fc/18h/"
outpath="/project/haggis/ERA5/forincrements/forecast/t2m/18h_step12/"
```

```python
# open up CESM data to get the output grid.
cesmdat = xr.open_dataset("/project/cas02/islas/CESM2LE/mon/1850_1950/TREFHT/TREFHT_1850_1950_1001.001.nc")
grid_out = xr.Dataset({'lat': (['lat'], cesmdat.lat)}, {'lon': (['lon'], cesmdat.lon)})
```

```python
reusewgt=False
wgtfile="/project/cas/islas/temp/era5/wgtfile.nc"
for iyear in np.arange(ystart,yend+1,1):
    print(iyear)
    dat = xr.open_dataset(inpath+"/tas_fc_18h_step12_"+str(iyear)+".nc")
    regridder = xe.Regridder(dat, grid_out,'bilinear', periodic=True, reuse_weights=reusewgt, filename=wgtfile)
    dat_rg = regridder(dat["2T_GDS4_SFC"])
    dat_rg = dat_rg.rename({"initial_time0_hours":"time"})
    dat_rg = dat_rg.rename("t2m")
    dat_rg.to_netcdf(outpath+"/tas_fc_18h_step12_"+str(iyear)+".nc")
```
