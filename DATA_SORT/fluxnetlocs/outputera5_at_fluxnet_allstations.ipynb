```python
import importlib
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys

from glob import glob
import os
```

```python
fnetdatadir="/project/cas/islas/python_savs/snowpaper/DATA_SORT/fluxnet/"
pathout="/project/cas/islas/python_savs/snowpaper/DATA_SORT/fluxnetlocs_era5/"
datdir="/project/mojave/observations/ERA5_daily/"
```

```python
# read in the stations
stations = [os.path.basename(x) for x in glob(fnetdatadir+'/*')]
for istation in stations:
    name  = istation[0:6]
    print(name)
    fname = fnetdatadir+'/'+istation
    data = xr.open_dataset(fname)
    
    lonstat = data.lon.values
    latstat = data.lat.values
    
    if (lonstat < 0):
        lonstat =360+lonstat
        

    era5_trefht = xr.open_mfdataset(datdir+"T2m/*.nc",
                                   coords='minimal', join='override', decode_times=True).sel(lon=lonstat, lat=latstat, method='nearest')
    era5_trefht = era5_trefht.sel(time=slice("1979-01-01","2014-12-31"))
    era5_trefht= era5_trefht.rename(t2m='era5_trefht')
    era5_trefht.to_netcdf(path=pathout+istation)
    del era5_trefht
    print('done TREFHT')
    
    era5_shflx = xr.open_mfdataset(datdir+"SHFLX/*.nc",
                                  coords='minimal', join='override', decode_times=True).sel(lon=lonstat, lat=latstat, method='nearest')
    era5_shflx = era5_shflx.sel(time=slice("1979-01-01","2014-12-31"))
    era5_shflx = era5_shflx.rename(shflx='era5_shflx')
    era5_shflx.to_netcdf(path=pathout+istation, mode="a")
    del era5_shflx
    print('done SHFLX')
    
    era5_flns = xr.open_mfdataset(datdir+"STR/*.nc",
                                 coords='minimal', join='override', decode_times=True).sel(lon=lonstat, lat=latstat, method='nearest')
    era5_flns = era5_flns.sel(time=slice("1979-01-01","2014-12-31"))
    era5_flns = era5_flns.rename(STR='era5_flns')
    era5_flns.to_netcdf(path=pathout+istation, mode="a")
    del(era5_flns)    
    print('done FLNS')
    
    era5_fsns = xr.open_mfdataset(datdir+"SSR/*.nc",
                                 coords='minimal', join='override', decode_times=True).sel(lon=lonstat, lat=latstat, method='nearest')
    era5_fsns = era5_fsns.sel(time=slice("1979-01-01","2014-12-31"))
    era5_fsns = era5_fsns.rename(SSR='era5_fsns')
    era5_fsns.to_netcdf(path=pathout+istation, mode="a")
    del(era5_fsns)        
    print('done FSNS')
    
    era5_lhflx = xr.open_mfdataset(datdir+"LHFLX/*.nc",
                                  coords='minimal', join='override', decode_times=True).sel(lon=lonstat, lat=latstat, method='nearest')
    era5_lhflx = era5_lhflx.sel(time=slice("1979-01-01","2014-12-31"))
    era5_lhflx = era5_lhflx.rename(lhflx='era5_lhflx')
    era5_lhflx.to_netcdf(path=pathout+istation, mode="a")
    del era5_lhflx
    print('done LHFLX')
```
