```python
plotpath="/project/cas/islas/python_plots/snowpaper/FIGURES/"
```

```python
basepath="/project/cas/islas/python_savs/snowpaper/DATA_SORT/flux_trefht_gradients/"
clm5 = xr.open_dataset(basepath+'gradients_Isla_CAM6_CLM5_002.nc')
snowd = xr.open_dataset(basepath+'gradients_CAM6_CLM5_snowdensity_002.nc')
```

Make our regions that have FSNO < 0.5 or are not land

```python
landfrac = xr.open_dataset('/project/cas/islas/cesmle/fx/landfraclens.nc')
landfrac = landfrac.LANDFRAC.isel(time=0)

path="/project/cas02/islas/CLM5_CLM4/raw/Isla_CAM6_CLM5_002/mon/lnd/"
fsno_clm5 = read.read_sfc_cesm(path+"FSNO_f.e21.FHIST.f09_f09.cesm2_cam6_clm5.002.1979-01_2014-12.nc","1979-01","2014-12")
fsno_clm5_djf = cal.season_mean(fsno_clm5,"FSNO",season="DJF")

path="/project/cas02/islas/CLM5_CLM4/raw/CAM6_CLM5_snowdensity_002/mon/lnd/"
fsno_snowd = read.read_sfc_cesm(path+"FSNO_f.e21.FHIST_BGC.f09_f09.cesm2_cam6_clm5_snowdensity.002.1979-01_2014-12.nc","1979-01","2014-12")
fsno_snowd_djf = cal.season_mean(fsno_snowd,"FSNO", season="DJF")

maskclm5 = np.empty([snowd.lat.size, snowd.lon.size])
maskclm5[ fsno_clm5_djf > 0.5] = 1
maskclm5[ landfrac < 0.5] = 0

masksnowd = np.empty([snowd.lat.size, snowd.lon.size])
masksnowd[ fsno_snowd_djf > 0.5] = 1
masksnowd[ landfrac < 0.5 ] = 0

graymaskclm5 = np.zeros([snowd.lat.size, snowd.lon.size])*nan
graymaskclm5[ fsno_clm5_djf < 0.5 ] = 1
graymaskclm5[ landfrac < 0.5] = nan

graymasksnowd = np.zeros([snowd.lat.size, snowd.lon.size])*nan
graymasksnowd[ fsno_snowd_djf < 0.5 ] = 1
graymasksnowd[ landfrac < 0.5] = nan
```

```python
clm5 = clm5*np.array(landfrac)*maskclm5
snowd = snowd*np.array(landfrac)*masksnowd
```

```python
fig = plt.figure(figsize=(16,16))

ax1 = maps.contourmap_bothcontinents_fill_nh_pos(fig, np.array(clm5.bsumflux) - np.array(snowd.bsumflux),
                                                np.array(snowd.lon), np.array(snowd.lat),0.02,-1.6,1.6,
                                                '(a) F$\\uparrow$ vs T$_{2m}$ gradient, CLM5$-$SNWDENS',
                                                0.05,0.32,0.8,0.95)
dat, lon = add_cyclic_point(graymaskclm5, coord=clm5.lon)
ax1.contourf(lon, np.array(clm5.lat), dat, levels=[0.99,1.01], colors=['lightgray','lightgray'])


ax1 = maps.contourmap_bothcontinents_fill_nh_pos(fig, np.array(clm5.bshflx) - np.array(snowd.bshflx),
                                                np.array(snowd.lon), np.array(snowd.lat),0.02,-1.6,1.6,
                                                 '(b) SH$\\uparrow$ vs T2m gradient, CLM5$-$SNWDENS',
                                                 0.36,0.63,0.8,0.95)
dat, lon = add_cyclic_point(graymasksnowd, coord=snowd.lon)
ax1.contourf(lon, np.array(snowd.lat), dat, levels=[0.99,1.01], colors=['lightgray','lightgray'])

ax1 = maps.contourmap_bothcontinents_fill_nh_pos(fig, np.array(clm5.bflns) - np.array(snowd.bflns),
                                                np.array(snowd.lon), np.array(snowd.lat),0.02,-1.6,1.6,
                                                '(c) LW$_{net}\\uparrow$ vs T2m gradient, CLM5$-$SNWDENS',
                                                 0.67,0.94,0.8,0.95)
dat, lon = add_cyclic_point(graymaskclm5, coord=clm5.lon)
ax1.contourf(lon, np.array(clm5.lat), dat, levels=[0.99,1.01], colors=['lightgray','lightgray'])


ax = cbars.plotcolorbar(fig, 0.02,-1.6,1.6,'Gradient (Wm$^{-2}$K$^{-1}$)',
                       0.25,0.74,0.76,0.77,orient='horizontal',
                       ticks=[-1.5,-1,-0.5,0,0.5,1,1.5], fsize=12)

fig.savefig(plotpath+'fig9.pdf', facecolor='white', bbox_inches='tight')
```
